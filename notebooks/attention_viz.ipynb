```python
import torch
from pcnn.models.pcnn import PCNN
from pcnn import DATA_DIR
import os
from hydra import compose, initialize
import hydra

import matplotlib.pyplot as plt
```

```python
dir_exp = os.path.join(DATA_DIR,"..","logs","experiments","multiruns","legs_modelnet_learnt","2023-09-05_12-05-54","1")
dir_exp_rel = os.path.join("..","logs","experiments","multiruns","legs_modelnet_learnt","2023-09-05_12-05-54","1")
model = PCNN.load_from_checkpoint(os.path.join(dir_exp,"tmp","epoch=71-step=7200.ckpt"))
```

```python
cfg_dir = os.path.join(dir_exp_rel,".hydra")
with initialize(version_base=None, config_path=cfg_dir):
    cfg = compose(config_name="config.yaml")
```

```python
datamodule = hydra.utils.instantiate(cfg.data)()
datamodule.prepare_data()
```

```python
for i,b in enumerate(datamodule.test_dataloader()):
    break
```

```python
graph_out = model.forward(b)

pooled, probs = model.pooling(graph_out,return_probs=True)
```

```python
idx = 6
graph_ = graph_out.to_data_list()[idx].detach()
prob = probs[idx].detach()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
sc = ax.scatter(graph_.pos[:,0],graph_.pos[:,1],graph_.pos[:,2],c=prob[:,0])
plt.colorbar(sc)
ax.set_title("Pooling Attention")
plt.show()
```
